```python
# Import necessary libraries 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs, make_circles
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering

%matplotlib inline 
```

```python
#function to display plot clustering results 

def plot_clusters(X, labels, ax, title):
    scatter = ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolor='k')
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Feature 1', fontsize=12)
    ax.set_ylabel('Feature 2', fontsize=12)
    # Generate legend handles and labels directly from the scatter object
    handles, legend_labels = scatter.legend_elements()
    ax.legend(handles, legend_labels, title="Clusters", fontsize=10)
```


# Dataset 1: make_moons

```python
# Generate the moons dataset
Xmoons, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Create a figure with 3 subplots for the three clustering methods
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Apply DBSCAN
dbs_moons = DBSCAN(eps=0.3, min_samples=5)
labs_dbscan = dbs_moons.fit_predict(Xmoons)
plot_clusters(Xmoons, labs_dbscan, axs[0], 'DBSCAN Clustering on Moons Dataset')

# Apply k-Means (using 2 clusters since moons data has two distinct shapes)
km_moons = KMeans(n_clusters=2, random_state=42)
labs_kmeans = km_moons.fit_predict(Xmoons)
plot_clusters(Xmoons, labs_kmeans, axs[1], 'k-Means Clustering on Moons Dataset')

# Apply Hierarchical Clustering (Agglomerative)
agg_moons = AgglomerativeClustering(n_clusters=2)
labs_agg = agg_moons.fit_predict(Xmoons)
plot_clusters(Xmoons, labs_agg, axs[2], 'Hierarchical Clustering on Moons Dataset')

#display plots

plt.tight_layout()
plt.show()
```

# Dataset 2: make_blobs 

```python
# Generate the blobs dataset with three centers and varying standard deviations
Xblobs, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Apply DBSCAN
dbs_blobs = DBSCAN(eps=2.0, min_samples=5)
labs_dbscan = dbs_blobs.fit_predict(Xblobs)
plot_clusters(Xblobs, labs_dbscan, axs[0], 'DBSCAN Clustering on Blobs Dataset')

# Apply k-Means (3 clusters)
kms_blobs = KMeans(n_clusters=3, random_state=42)
labs_kmeans = kms_blobs.fit_predict(Xblobs)
plot_clusters(Xblobs, labs_kmeans, axs[1], 'k-Means Clustering on Blobs Dataset')

# Apply Hierarchical Clustering (Agglomerative with 3 clusters)
agg_blobs = AgglomerativeClustering(n_clusters=3)
labs_agg = agg_blobs.fit_predict(Xblobs)
plot_clusters(Xblobs, labs_agg, axs[2], 'Hierarchical Clustering on Blobs Dataset')

plt.tight_layout()
plt.show()
```

# Dataset 3: make_circles 

```python
# Generate the circles dataset
Xcircles, _ = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Apply DBSCAN
dbscan_circles = DBSCAN(eps=0.2, min_samples=5)
labs_dbscan = dbscan_circles.fit_predict(Xcircles)
plot_clusters(Xcircles, labs_dbscan, axs[0], 'DBSCAN Clustering on Circles Dataset')

# Apply k-Means (2 clusters)
kms_circles = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kms_circles.fit_predict(Xcircles)
plot_clusters(Xcircles, labels_kmeans, axs[1], 'k-Means Clustering on Circles Dataset')

# Apply Hierarchical Clustering (Agglomerative with 2 clusters)
agg_circles = AgglomerativeClustering(n_clusters=2)
las_agg = agg_circles.fit_predict(Xcircles)
plot_clusters(Xcircles, las_agg, axs[2], 'Hierarchical Clustering on Circles Dataset')

plt.tight_layout()
plt.show()
```
